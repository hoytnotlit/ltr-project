```python
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse('leo_coctaill.xml')
root = tree.getroot()
clean_data = [" ".join([word.text for word in sent]) for sent in root]
```

```python
with open('corrupted.txt', 'r') as corr, open('corrupted_data.tsv', 'w') as comp:
    for line in corr:
        # include:
        # index of original sentence from leo_coctaill.xml converted to python list
        # original sentence
        # corrupted sentence 
        # removed word 
        # position in corrupted sentence where removed word should be 
        corr_line = line.split('\t')
        orig_i = int(corr_line[0])
        orig_s = clean_data[orig_i]
        corr_s = corr_line[1].replace("\n", "")
        missing_w = ""
        pre_w_i = None
        for i in range(len(corr_s.split())):
            if orig_s.split()[i] != corr_s.split()[i]:
                missing_w = orig_s.split()[i]
                pre_w_i = i
                break
            
        l = [str(orig_i), orig_s, corr_s, missing_w, str(pre_w_i)]
        print(l)
        comp.write('\t'.join(l) + '\n')
```
